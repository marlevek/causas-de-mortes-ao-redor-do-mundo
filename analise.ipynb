```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Importando o conjunto de dados e lendo as 10 primeiras linhas
dados = pd.read_csv('cause_of_deaths.csv')
dados.head(10)
```

```python
# morte por malária no brasil ao longo dos anos
brasil = dados[dados['Country/Territory'] == 'Brazil']

plt.figure(figsize=(10, 6))
plt.plot(brasil['Year'], brasil['Malaria'], marker='o')
plt.xlabel('Ano')
plt.ylabel('Óbitos de Malária')
plt.title('Óbitos de Malária no Brasil ao Longo dos Anos')
plt.grid(True)
plt.show()
```

```python
# morte por deficiência nutricional no brasil ao longo dos anos
plt.figure(figsize=(10, 6))
plt.plot(brasil['Year'], brasil['Nutritional Deficiencies'], marker='o')
plt.xlabel('Ano')
plt.ylabel('Óbitos por Deficiência Nutricional')
plt.title('Óbitos por Deficiência Nutricional ao longo dos anos')
plt.grid(True)
plt.show();
```

5 Principais causas de morte no brasil devido a doenças

```python
# Selecione as colunas de causas de morte (ignorando as colunas 'Country/Territory', 'Code' e 'Year')
causas_de_morte = brasil.drop(columns=['Country/Territory', 'Code', 'Year'])

# Calcule a soma total de mortes para cada causa
somas_causas = causas_de_morte.sum().sort_values(ascending=False)

# Selecione as 5 maiores causas
top_5_causas_brasil = somas_causas.head(5)

# Plote um gráfico de barras
plt.figure(figsize=(12, 6))
top_5_causas_brasil.plot(kind='bar')
plt.xlabel('Causas de Morte')
plt.ylabel('Número Total de Mortes')
plt.title('5 Maiores Causas de Morte no Brasil')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show();
```

```python
usa = dados[dados['Country/Territory'] == 'United States']


causas_de_morte = usa.drop(columns=['Country/Territory', 'Code', 'Year'])

# Calcule a soma total de mortes para cada causa
somas_causas = causas_de_morte.sum().sort_values(ascending=False)

# Selecione as 5 maiores causas
top_5_causas_usa = somas_causas.head(5)

# Plote um gráfico de barras
plt.figure(figsize=(12, 6))
top_5_causas_usa.plot(kind='bar')
plt.xlabel('Causas de Morte')
plt.ylabel('Número Total de Mortes')
plt.title('5 Maiores Causas de Morte nos Estados Unidos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show();
```

```python
# Combine os dados das 5 maiores causas em cada país
combined_causas = pd.DataFrame({
    'Brasil': top_5_causas_brasil,
    'USA': top_5_causas_usa
}).fillna(0)

# Plote um gráfico de barras empilhadas
combined_causas.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Causas de Morte')
plt.ylabel('Número Total de Mortes')
plt.title('Comparação das 5 Maiores Causas de Morte no Brasil e nos EUA')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```
